Let's gather the average transactions per month.

```python
lowest_amount = df1[df1['amount'] == df1['amount'].min()]
lowest_amount.shape
```

```python
lowest_amount
```

There are 26 cases in which the lowest amount is 0.10 AUD.

```python
df1['month'] = df1.date.dt.month
df1.head(3)
```

```python
df1.groupby('month').mean()
```

Three months are represented for this dataset. The monthly average balance ranged from 10794AUD to 18461AUD, and the transaction amount was typically 182AUD to 196AUD.

```python
df1['day_of_week'] = df1.date.dt.day_name()
df1.head(3)
```

```python
df1.groupby('day_of_week').mean()
```

The account balance and age are fairly consistent across all days. However, the transaction amount varies greatly. The lowest average transactions were done on Sundays at 53AUD, and the highest was on Mondays at 373AUD.

```python
df1['extraction_time'] = pd.to_datetime(df1['extraction'], format='%Y-%m-%d %H:%M:%S').dt.time
df1.head(3)
```

```python
type(df1['extraction_time'][0])
```

```python
df1['extraction_hour'] = pd.to_datetime(df1['extraction'], format='%Y-%m-%d %H:%M:%S').dt.hour
df1.head(3)
```

```python
df1.tail(3)
```

```python
plt.bar(df1['extraction_hour'], df1['amount'])
plt.xlabel('Hour of Transaction (h)')
plt.ylabel('Amount of Transaction (AUD)')
plt.title('Transaction Amount per Hour for ANZ')
```

The highest volume of transactions (more than 8000AUD) were done at 1PM.
